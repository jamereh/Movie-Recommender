```python
df2 = pd.read_csv('tmdb_5000_movies.csv')
df2.head(20)
```

```python
df1.columns = ['id', 'tittle', 'cast', 'crew']
df2 = df2.merge(df1, on = 'id')
```

```python
df2.head(20)
```

```python
#average vote of all movies. Avg mean is about 6 on a scale of 10
C = df2['vote_average'].mean()
C
```

```python
#Determining which movies qualifify to be featured on recommended list. m is the min amount of votes required to be featured on the list
m = df2['vote_count'].quantile(0.8) #only movies with more votes that 80% of all movies on list will be featured
m
```

```python
#Filtering out movies that qualify
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies
```

```python
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#Defining a score column and calculating its value with weighted rating function
q_movies['score'] = q_movies.apply(weighted_rating, axis = 1)
q_movies
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

q_movies[['title', 'vote_count', 'vote_average', 'score']].head(15)
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
df2['overview'] = df2['overview'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df2['overview'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
#Use linear kernel since I have used the TF-IDF vectorizer. 
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

#Computing consine similarity matrix using linear kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
```

```python
# Function that takes in movie title as input and outputs most similar movies
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df2['title'].iloc[movie_indices]
```

```python
#The recommender is doing an okay job recommending movies, but I think it could be better. Next, I'm going to build a recommender that makes suggestions based on credits, genres, crew, etc.

from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
#I'm creating a few functions that will extract information from each feature

def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
#return list of the top three elements

def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    
    return []
```

```python
#define new director, genres, cast and keywords features

df2['director'] = df2['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
#print the new features
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Apply clean_data function to your features.
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

```python
# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

```python
get_recommendations('The Dark Knight Rises', cosine_sim2)
```

```python
get_recommendations('The Godfather', cosine_sim2)
```

```python
#So far, the recommenders built have only been able to suggest movies similar to a certain movie. They aren't personal and anyone using the recommender will receive the same suggestions. 

#Next I will use collaborative filtering to build a recommender that will make more personal recommendations
```

```python
from surprise import Reader, Dataset, SVD
reader = Reader(rating_scale=(1, 5))
ratings = pd.read_csv("C:/Users/jamer/Movie Recommender/ratings_small.csv")
ratings.head()
```

```python
algorithm = SVD()
```

```python
from surprise.model_selection import cross_validate

data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
results = cross_validate(algorithm, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
trainset = data.build_full_trainset()
algorithm.fit(trainset)
```
